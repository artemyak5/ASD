<p style="font-size: 15x;"><strong> 1.1. Теоретичні відомості </strong></p>

<p style="font-size: 15px;"> Рекурсія відноситься до одного з фундаментальних понять у математичних та комп’ютерних науках. В мовах програмування рекурсивною програмою називають програму, яка звертається сама до себе (сама себе викликає). Рекурсивна програма не може викликати себе до нескінченності, оскільки у цьому випадку вона ніколи не завершиться. Другою важливою особливістю рекурсивної програми є умова завершення, що дозволяє програмі припинити себе викликати.</p>

<p style="font-size: 15px;"><strong>Принципові властивості рекурсивного визначення:</strong><br>
1. визначення об'єкту самого через себе, але з іншими параметрами; <br>2. завершеність ланцюжка визначень на деякому значенні аргументів.</p>
    
<p style="font-size: 15px;">Необхідність використання рекурсії виникає при реалізації динамічних
структур даних, таких як дерева, стеки, черги. Для реалізації рекурсивних
алгоритмів передбачена можливість створення рекурсивних функцій.
Рекурсивна функція являє собою функцію, у тілі якої здійснюється виклик
цієї ж функції.</p>

<p style="font-size: 15px;">Визначити, чи є задана послідовність чисел арифметичною прогресією за допомогою рекурсії.</p>

---
<p style="font-size: 15px;"><strong>Параметри вхідних даних</strong></p>

<p style="font-size: 15px;">seq_of_num: список чисел, який потрібно перевірити на арифметичну прогресію.</p>

---
<p style="font-size: 15px;"><strong>Параметри вихідних даних</strong></p>

<p style="font-size: 15px;">True, якщо послідовність є арифметичною прогресією.
False, якщо послідовність не є арифметичною прогресією або містить менше ніж 3 елементи.</p>

---

<p style="font-size: 15px;"><strong>Реалізація</strong></p>

<p style="font-size: 15px;">Перевіряється довжина послідовності. Якщо вона менше 3, повертається False, оскільки для арифметичної прогресії потрібно щонайменше 3 елементи.
Перевіряється, чи різниця між першим і другим елементами співпадає з різницею між другим і третім елементами. Якщо так, рекурсивно перевіряється решта послідовності.
Якщо всі перевірки успішні, повертається True.</p>

---

```python
def check_for_arefmatic_seq(seq_of_num):
    # Якщо в послідовності менше 3-х елементів, то неможливо визначити арифметичну прогресію
    if len(seq_of_num) < 3:
        return False
    # перевіряємо, чи різниця між 1-м і 2-м елементом співпадає з різницею між 2-м і 3-м елементами
    if seq_of_num[1] - seq_of_num[0] == seq_of_num[2] - seq_of_num[1]:

        if len(seq_of_num) == 3:
            return True
        
        return check_for_arefmatic_seq(seq_of_num[1:])
    else:
        return False
    

seq_of_num = [2, 4, 6, 8, 10, 12, 14, 16, 18]  

if check_for_arefmatic_seq(seq_of_num):
    print(' -- Послідовність є арефметичною прогресією :)')
else:
    print(' -- Послідовність не є арефметичною прогресією :(')
```

<p style="font-size: 15px;">Реалізувати алгоритм для розв’язання задачі «Ханойські вежі». <br>Виписати послідовність ходів для перекладання n дисків вежі (n = 2; 3; 4; 5 дисків, використати онлайн гру).</p>

<p style="font-size: 15px;"><strong>Параметри вхідних даних</strong></p>

<p style="font-size: 15px;"><strong>num_disks:</strong> кількість дисків в вежі.
<br><strong>start:</strong> назва вежі, з якої починається переміщення дисків.
<br><strong>end:</strong> назва вежі, на яку необхідно перемістити всі диски.
<br><strong>temp:</strong> назва проміжної вежі, яка використовується для переміщення дисків.</strong></p>

<p style="font-size: 15px;"><strong>Реалізація</strong></p>

<p style="font-size: 15px;">Якщо num_disks дорівнює 1, виводиться повідомлення про переміщення одного диска з вежі start на вежу end.
Викликається рекурсивно функція hanoi для переміщення num_disks - 1 дисків з вежі start на вежу temp.
Виводиться повідомлення про переміщення диска num_disks з вежі start на вежу end.
Знову викликається рекурсивно функція hanoi для переміщення num_disks - 1 дисків з вежі temp на вежу end.</p>

```python
def hanoi(num_disks, start, end, temp):
    if num_disks == 1:
        print("Перемістити диск 1 з {} на {}".format(start, end))
        return
    hanoi(num_disks - 1, start, temp, end)
    print("Перемістити диск {} з {} на {}".format(num_disks, start, end))
    hanoi(num_disks - 1, temp, end, start)

# Приклад використання для 3 дисків
num_disks = 3
hanoi(num_disks, 'Початковий', 'Кінцевий', 'Проміжний')
```
